```r
#转为数值型
table(Affairs$affairs)
a <- sub("female",1,Affairs$gender)
b <- sub("male",0,a)
Affairs$gender <- a                         
Affairs$gender <-b
as.numeric(Affairs$gender)

table(Affairs$children)
c <- sub("yes",1,Affairs$children)
d<- sub("no",0,c)
Affairs$children <- c                         
Affairs$children<-d
as.numeric(Affairs$children)
```

```r
Affairs$affairs2 = ifelse(Affairs$affairs == 0, 0 , 1)
```

```r
#建立模型
fit.full <- glm(affairs2 ~ gender+age+yearsmarried+children+religiousness+education+occupation+rating,
                data=Affairs,family = binomial())
summary(fit.full)
```
