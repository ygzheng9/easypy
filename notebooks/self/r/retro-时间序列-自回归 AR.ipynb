```r
# 随机游走的数据集
set.seed(0)

x <- rnorm(1000)       # 生成符合正态分布N(0,1)的数据
w <- x

for(t in 2:1000) { 
    x[t] <- x[t-1] + w[t]
}

tsx <- ts(x)              # 生成ts时间序列的数据集

# 查看数据集
head(tsx, 15)
```

```r
plot(tsx)            # 生成可视化图形
```

```r
a <- ar(tsx)        # 进行自回归建模
a
```

```r
b<-ar(tsx,method = "ols");b
```

```r
d<-ar(tsx,method = "mle");d
```
