```python
# 导入相关库
import pandas as pd
from pyecharts.charts import Page, Sankey
from pyecharts import options as opts

# 读取csv文件
data = pd.read_csv('./data/flow.csv', encoding='UTF-8', header=None, sep=' ')

# 生成nodes
nodes = []
nodes.append({'name': '总支出'})
for i in data[0].unique():
    dic = {}
    dic['name'] = i
    nodes.append(dic)

# 生成links
links = []
for i in data.values:
    dic = {}
    dic['source'] = i[0]
    dic['target'] = i[1]
    dic['value'] = i[2]
    links.append(dic)
```

```python
data[0]
nodes
data.values
links
```

```python
# pyecharts 所有方法均支持链式调用。
c = (Sankey().add(
    "费用/元",
    nodes,
    links,
    linestyle_opt=opts.LineStyleOpts(opacity=0.2,
                                     curve=0.5,
                                     color="source",
                                     type_="dotted"),
    label_opts=opts.LabelOpts(position="right", ),
).set_global_opts(title_opts=opts.TitleOpts(title="我的生活支出一览")))
# 输出html可视化结果
# c.render('result.html')
c.render_notebook()
```
