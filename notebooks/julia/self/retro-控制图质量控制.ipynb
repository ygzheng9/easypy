# 场景1：模拟控制图，围绕均值上下随机分布
1. 如果超过边界值(1个标准差)，视为异常；
2. 如果在均值线同侧连续 7 次，视为异常；
3. 还有很多模式，用以判定异常；

```julia
mu = 5.5
theta = 0.3
d = Normal(mu, theta)

cnt = 50
a = rand(d, cnt)
t = range(1,step=1, length=cnt)

gr()
plot(a, xtick=1:5:50, ylim=(4.5,6.5), legend=false)

# 增加水平线
plot!([ mu, mu - 2* theta, mu + 2* theta], linetype="hline", linestyle = :dash, linecolor = :red)
```
