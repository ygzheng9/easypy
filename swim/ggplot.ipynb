```python
from plotnine import *
from plotnine.data import mtcars

mtcars.head()
```

```python
# Visualize two-dimensions (2-D)

(ggplot(mtcars, aes('wt', 'mpg')) 
+ geom_point()
+ theme_bw())
```

```python
# Visualize three-dimensions (3-D)

(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
+ geom_point()
+ theme_bw())
```

```python
# Visualize four-dimensions (4-D)

(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)', size='cyl'))
+ geom_point() 
+ theme_bw())
```

```python
(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
+ geom_point() 
+ facet_wrap('~cyl') 
+ theme_bw())
```

```python
# Visualize data dimensions and statistics

(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
+ geom_point() 
+ stat_smooth(method='lm') 
+ theme_bw())
```

```python
# Visualize five-dimensions (5-D)

(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)', size='cyl'))
+ geom_point() 
+ facet_wrap('~am') 
+ theme_bw())
```

```python
# Visualize six-dimensions (6-D)

(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)', size='cyl'))
+ geom_point() 
+ facet_grid('am ~ carb') 
+ theme_bw())
```
