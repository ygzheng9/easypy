```python
# 电影、用户、评分
# 通过 join，在 评分表 中，显示 电影名字
# 按照 电影名字，计算 总评分次数，总评分，平均评分
# 按 总评分 从大到小排序

# 选定一部电影，比如：Forest Gun
# 计算 这部电影 和 其它电影的相似度

# 把 评分表 转成 矩阵形式（行：用户；列：电影）
# 计算 相似度，但是需要只计算 评分次数 大于 50 的电影的；
# 对相似度从大到小排序
```

```python
import numpy as np
import pandas as pd
```

```python
baseDir = '../dataset/ml-latest-small'

# 读入数据文件：评分表
ratings_data = pd.read_csv('%s/ratings.csv' % (baseDir) )
ratings_data.head()
```

```python
# 读入数据文件
movie_names = pd.read_csv('%s/movies.csv' % (baseDir) )  
movie_names.head() 
```

```python
# 通过 movieId 把两个 dataframe join
movie_data = pd.merge(ratings_data, movie_names, on='movieId')
movie_data.head()
```

```python
# 按 电影名 求 平均分，并且 排序
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
# 按 电影名 求 评分次数，并 排序
movie_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
# 新建一个 dataFrame，记录 电影，平均评分
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())

# 增加一列，评分的次数
ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
ratings_mean_count.head()
```

```python
# 准备画图
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
%matplotlib inline

plt.figure(figsize=(8,6))
plt.rcParams['patch.force_edgecolor'] = True

```

```python
# 评分次数的 直方图
ratings_mean_count['rating_counts'].hist(bins=50)
```

```python
# 平均分 的直方图
ratings_mean_count['rating'].hist(bins=50)
```

```python
# 平均评分 和 评分次数 的联合
# 相对于 低分电影(比如：低于 2.5)，高分电影（3.5 ~ 4） 的 评论次数 更多
sns.jointplot(x='rating', y='rating_counts', data=ratings_mean_count, alpha=0.4)
```

```python
# 之下是协同过滤

# 做透视图，行：用户；列：电影；中间单元格：用户对该电影的评价
user_movie_rating = movie_data.pivot_table(index='userId', columns='title', values='rating')
```

```python
user_movie_rating.head()
```

```python
# 选中一部电影
forrest_gump_ratings = user_movie_rating['Forrest Gump (1994)']
forrest_gump_ratings.head()
```

```python
# 求 和这部电影 相似的电影
movies_like_forest_gump = user_movie_rating.corrwith(forrest_gump_ratings)

corr_forrest_gump = pd.DataFrame(movies_like_forest_gump, columns=['Correlation'])
corr_forrest_gump.dropna(inplace=True)
corr_forrest_gump.head()
```

```python
# 按 相关系数，排序
corr_forrest_gump.sort_values('Correlation', ascending=False).head(10)
```

```python
# 对 dataFrame 增加一列
corr_forrest_gump = corr_forrest_gump.join(ratings_mean_count['rating_counts'])
corr_forrest_gump.head()
```

```python
# 过滤掉 评论次数 小于 50 的电影
corr_forrest_gump[corr_forrest_gump ['rating_counts']>50].sort_values('Correlation', ascending=False).head(10)
```
