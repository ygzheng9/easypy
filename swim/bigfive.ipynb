```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# 二项分布
n = 10
p = 0.3

# 在每个点的概率
x = np.arange(0, 21)
binomial = stats.binom.pmf(x, n, p)
binomial
```

```python
print('n = %i , p = %.2f' % (n, p))
```

```python
plt.plot(x, binomial, 'o-')
plt.title('Binomial: n=%i , p=%.2f' % (n, p), fontsize=15)
plt.xlabel('Number of success')
plt.ylabel('Probability of success', fontsize=15)
plt.show()
```

```python
binom_sim = stats.binom.rvs(n=10, p=0.5, size=1000)

plt.hist(binom_sim, bins=10, normed=True)
plt.xlabel('x')
plt.ylabel('Density')
plt.show()
```

```python
# 模拟 20 次投硬币，正反面概率 50%
# 记下每一次的结果
a = [stats.binom.rvs(n=1, p=0.5, size=1) for i in range(20)]
for i in range(20): 
    print(f"({i}, {a[i][0]})")
```

```python
# 1 到 500 次
r = range(500)

# 基数
base = [ i for i in range(1, 501) ]

# 1 到 6 个硬币
# 每个硬币 500 次，每次的结果
for t in range(1, 6):
    # 每一个硬币，模拟抛 500 次，返回值是 每一次的正反
    a = [stats.binom.rvs(n=1, p=0.5, size=1) for i in r]
    b = [ a[i][0] for i in r]
    
    # 计算累计的正面数
    c = np.cumsum(b)
    
    # 计算累计正面的比例
    d = np.divide(c, base)
    
    print(f"{t} {d}")
```
